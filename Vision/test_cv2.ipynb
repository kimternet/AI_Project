```python
import cv2
from ultralytics import YOLO

#   
model = YOLO('D:/YOLO4/best.pt')

#   
cap = cv2.VideoCapture(0)

if not cap.isOpened():
    print("Error: Could not open video stream.")
    exit()

#  
threshold = 0.6

while True:
    #  
    ret, frame = cap.read()
    if not ret:
        break
    
    # YOLO   
    results = model.predict(frame)

    #    
    annotated_frame = frame.copy()
    
    for result in results:
        boxes = result.boxes.xyxy.numpy()  #  bounding box 
        scores = result.boxes.conf.numpy()  #  confidence scores
        class_ids = result.boxes.cls.numpy()  #   ID
        labels = model.names  #   (result.names  model.names )
        
        for box, score, class_id in zip(boxes, scores, class_ids):
            if score >= threshold:
                x1, y1, x2, y2 = map(int, box)
                label = labels[int(class_id)]
                cv2.rectangle(annotated_frame, (x1, y1), (x2, y2), (255, 0, 0), 2)
                cv2.putText(annotated_frame, f'{label} {score:.2f}', (x1, y1 - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.9, (255, 0, 0), 2)
    
    #  
    cv2.imshow('YOLO Detection', annotated_frame)
    
    # 'q'   
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

#  
cap.release()
cv2.destroyAllWindows()
```
